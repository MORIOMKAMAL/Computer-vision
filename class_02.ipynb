```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread("/content/images.jpeg")
plt.imshow(image)
plt.show()

print("Shape of the image : ", image.shape)

# Convert BGR to RGB for display
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Apply Average Filter (blur)
average_filtered = cv2.blur(image , (5,5))

# Convert filtered image to RGB
average_filtered_rgb = cv2.cvtColor(average_filtered, cv2.COLOR_BGR2RGB)

# Show the filtered image
plt.imshow(average_filtered_rgb)
plt.title("Average Filtered Image")
plt.show()


# Plot original and average filtered images
plt.figure(figsize=(10,5))

plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(average_filtered_rgb)
plt.title('Average (Mean) Filtered Image')
plt.axis('off')

plt.show()

# cv2.GaussianBlur(src, ksize, sigmaX)
gaussian_filtered = cv2.GaussianBlur(image_rgb, (5, 5), 0)

# Plot original and Gaussian filtered images
plt.figure(figsize=(10,5))

plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(gaussian_filtered)
plt.title('Gaussian Filtered Image')
plt.axis('off')

plt.show()

# cv2.medianBlur(src, ksize)
median_filtered = cv2.medianBlur(image_rgb, 5)

# Plot original and Gaussian filtered images
plt.figure(figsize=(10,5))

plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(median_filtered)
plt.title('Median Filtered Image')
plt.axis('off')

plt.show()

# cv2.bilateralFilter(src, d, sigmaColor, sigmaSpace)
bilateral_filtered = cv2.bilateralFilter(image_rgb, d=9, sigmaColor=75, sigmaSpace=75)

plt.figure(figsize=(10,5))

plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(bilateral_filtered)
plt.title('Bilateral Filtered Image')
plt.axis('off')

plt.show()


```
